#Project Overview
This project implements a Content-Based Recommendation System using NLP techniques. The goal is to recommend similar items (like movies or products) based on their textual descriptions using TF-IDF and Cosine Similarity.

#Objectives
1. Load and preprocess a real-world dataset using Pandas.
2. Clean text data by removing stopwords, punctuation, and converting to lowercase.
3. Convert text into numerical vectors using TF-IDF Vectorization.
4. Compute Cosine Similarity to find the mathematical distance between items.
5. Build an interactive UI using Streamlit and deploy it on Render.

#Technical Stack
1. Language: Python 3
2. Environment: Google Colab / Local IDE
3. Libraries: * Pandas: For data manipulation.
     * Scikit-Learn: For TF-IDF and Cosine Similarity.
     * NLTK: For text preprocessing.
     * Streamlit: For building the web interface.
4. Deployment: Render Cloud Platform.

#Implementation Details
1. Data Preprocessing: Handled missing values by replacing them with empty strings. Applied a cleaning pipeline to the text column to ensure the model focuses on meaningful words.

2. Vectorization (TF-IDF): Used TfidfVectorizer with optimized ngram_range to capture both single words and pairs of words.

3. Similarity Computation: Generated a similarity matrix using Cosine Similarity, which measures the cosine of the angle between two vectors.

4. Recommendation Logic: Created a recommend() function that fetches the top 5 most similar items based on the similarity score.
